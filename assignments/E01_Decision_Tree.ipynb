# Questão 1

Utilize o dataset Iris disponível no scikit-learn.
Divida os dados em treino e teste utilizando divisão estratificada.

**Solução**:

```python
from sklearn import datasets
iris = datasets.load_iris()

from sklearn import datasets
import matplotlib.pyplot as plt
iris = datasets.load_iris()
from sklearn.model_selection import train_test_split

X = iris.data[:, :2]  # we only take the first two features.
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    test_size=0.3,
    random_state=42, 
    stratify=y          # divisão estratificada
)

plt.figure(),
plt.scatter(
    X_train[:,0], X_train[:,1],
    c=y_train,
    marker='o',
    label="Treino"
)

# test
plt.scatter(
    X_test[:,0], X_test[:,1],
    c=y_test,
    marker='x',
    label="Teste"
)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

# Questão 2

Treine um modelo utilizando `DecisionTreeClassifier`.

Depois calcule:

- acurácia no treino
- acurácia no teste

**Solução**:

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

iris = datasets.load_iris()

X = iris.data[:, :2]
y = iris.target

# divisão estratificada
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    stratify=y,
    random_state=42 # seed
)

# treinar modelo
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# acurácia
train_accuracy = model.score(X_train, y_train)
test_accuracy = model.score(X_test, y_test)

print("Acurácia no treino:", train_accuracy)
print("Acurácia no teste:", test_accuracy)
```

# Questão 3

Utilize `plot_tree()` para visualizar a árvore treinada.

Responda:

1. Qual atributo aparece na raiz?
2. Qual é a profundidade da árvore?

**Solução**:

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

iris = datasets.load_iris()

X = iris.data[:, :2]
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    stratify=y,
    random_state=42
)

model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

plt.figure(figsize=(10,6))
plot_tree(
    model,
    feature_names=iris.feature_names[:2],
    class_names=iris.target_names,
    filled=True
)
#plt.show()

plot_tree(model);# plt.show()

print(iris.feature_names[model.tree_.feature[0]]) # 1) Sepal Length (cm)

print(model.get_depth()) # 2) Profundidade é 11
```

# Questão 4

Treine dez árvores com:

- max_depth = 1
- max_depth = 2
- max_depth = 3
...
- max_depth = 9
- max_depth = None

Registre em uma tabela para cada árvore:

- acurácia no treino
- acurácia no teste
- profundidade da árvore
- número de folhas

**Solução**:

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

iris = datasets.load_iris()

X = iris.data[:, :2]
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    stratify=y,
    random_state=42
)

depths = [1,2,3,4,5,6,7,8,9,None]

results = []

for d in depths:
    model = DecisionTreeClassifier(max_depth=d, random_state=42)
    model.fit(X_train, y_train)

    results.append({
        "max_depth": d,
        "train_accuracy": model.score(X_train, y_train),
        "test_accuracy": model.score(X_test, y_test),
        "tree_depth": model.get_depth(),
        "num_leaves": model.get_n_leaves()
    })

table = pd.DataFrame(results)
print(table)

#   max_depth  train_accuracy  test_accuracy  tree_depth  num_leaves
#0        1.0        0.638095       0.600000           1           2
#1        2.0        0.800000       0.711111           2           4
#2        3.0        0.838095       0.733333           3           7
#3        4.0        0.838095       0.600000           4           9
#4        5.0        0.857143       0.733333           5          12
#5        6.0        0.876190       0.666667           6          15
#6        7.0        0.885714       0.644444           7          20
#7        8.0        0.914286       0.644444           8          26
#8        9.0        0.923810       0.644444           9          29
#9        NaN        0.942857       0.644444          11          33

#Então, a partir do depth 3, percebemos que o teste oscila bastante
```

# Questão 5

Treine dois modelos:

- criterion = "gini"

Compare:

- profundidade da árvore
- acurácia

**Solução**:

```python
from sklearn.tree import DecisionTreeClassifier

# modelo com gini
model_gini = DecisionTreeClassifier(criterion="gini", random_state=42)
model_gini.fit(X_train, y_train)

# modelo com entropy
model_entropy = DecisionTreeClassifier(criterion="entropy", random_state=42)
model_entropy.fit(X_train, y_train)

print("GINI")
print("Profundidade:", model_gini.get_depth())
print("Acurácia treino:", model_gini.score(X_train, y_train))
print("Acurácia teste:", model_gini.score(X_test, y_test))
print("ENTROPY")
print("Profundidade:", model_entropy.get_depth())
print("Acurácia treino:", model_entropy.score(X_train, y_train))
print("Acurácia teste:", model_entropy.score(X_test, y_test))
```

# Questão 6

Escolha um hiperparâmetro e investigue seu impacto.

Sugestões:

- max_depth
- min_samples_split
- min_samples_leaf
- criterion

Mostre resultados e interprete.
- melhor modelo encontrado
- acurácia
- parâmetros

**Solução**:

```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

values = [1,2,3,4,5,10]

results = []

for v in values:
    model = DecisionTreeClassifier(
        min_samples_leaf=v,
        random_state=42
    )

    model.fit(X_train, y_train)

    results.append({
        "min_samples_leaf": v,
        "train_accuracy": model.score(X_train, y_train),
        "test_accuracy": model.score(X_test, y_test),
        "depth": model.get_depth(),
        "leaves": model.get_n_leaves()
    })

table = pd.DataFrame(results)
print(table)

#- melhor modelo encontrado
#- acurácia
#- parâmetros

#O hiperparâmetro min_samples_leaf controla a complexidade da árvore. Valores pequenos permitem árvores profundas, 
#que tendem a sofrer overfitting. À medida que o valor aumenta, a árvore se torna mais simples. No experimento, 
#o melhor resultado foi obtido com min_samples_leaf = 10, que apresentou a maior acurácia no conjunto de teste (0.6667) 
#com uma árvore mais rasa (profundidade 4).
```
